```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from matplotlib import style; style.use('ggplot')
```

```python
# load data
data = pd.read_csv('exam_scores.csv')
data.head()
```

```python
# set up features and labels
X = data.values[:, [0, 1]]
y = data.values[:, -1]
m = y.size
```

```python
def plot_data():
    """ Plots data """
    # admitted: green, denied: red
    for i in range(m):
        if y[i] == 1: plt.scatter(X[i, 0], X[i, 1], color='green')
        if y[i] == 0: plt.scatter(X[i, 0], X[i, 1], color='red')
    
    # label and show plot
    plt.title('Exam Scores vs Admittance')
    plt.xlabel('Exam 1 Score')
    plt.ylabel('Exam 2 Score')
    
    plt.show()
    
    return None
```

```python
plot_data()
```

```python
def decision_boundary(clf):
    """ Plots decision boundary given a classifier """
    # entire grid
    X1, X2 = np.meshgrid(np.arange(start=min(X[:, 0]), stop=max(X[:, 0]), step=0.1),
                         np.arange(start=min(X[:, 1]), stop=max(X[:, 1]), step=0.1))
    
    # plot predictions
    plt.contourf(X1, X2, clf.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
                 alpha=0.75, cmap=ListedColormap(('lightpink', 'lightgreen')))
    
    # show original data on top
    plot_data()
    
    return None
```

# Your turn! Make some classifiers and test them out.
In every cell, you should:
- import the classifier from sklearn
- create an instance of the classifier
- fit the classifier
- make predictions with the decision boundary function

```python
# make KNN classifier
from sklearn.neighbors import KNeighborsClassifier
""" test out different numbers of neighbors: 1, 2, 25 """
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X, y)
decision_boundary(knn)
```

```python
# make logistic regression classifier
from sklearn.linear_model import LogisticRegression
""" test out different C values: 1, 3, 10 """
log_reg = LogisticRegression(C=100, solver='lbfgs')
log_reg.fit(X, y)
decision_boundary(log_reg)
```

```python
# make SVM classifier
from sklearn.svm import SVC
""" test out differnt polynomial degrees: 2, 1 """
svm = SVC(kernel='poly', degree=2, gamma='scale')
svm.fit(X, y)
decision_boundary(svm)
```

```python
# make decision tree classifier
from sklearn.tree import DecisionTreeClassifier
""" test out different max depths: None, 1, 2"""
tree = DecisionTreeClassifier(max_depth=None)
tree.fit(X, y)
decision_boundary(tree)
```

```python
# make neural net
from sklearn.neural_network import MLPClassifier
""" just run a bunch of times """
nn = MLPClassifier(hidden_layer_sizes=(25,), activation='relu', max_iter=5000)
nn.fit(X, y)
decision_boundary(nn)
```
